```python
# import packages
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import json
import numpy as np
import re
```

```python
# Scrape homepage for individual episode links
base_url = "https://bigbangtrans.wordpress.com/"
response = requests.get(base_url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# list of episode urls
episode_urls = []
r = re.compile(r'.*episode.*')

# find all links and append to list
for a in soup.find_all('a', href=True):
    episode_urls.append(a['href'])

# Filter for episode links
episode_urls = list(filter(r.match, episode_urls))
```

```python
# Download transcipts and add to the dictionary by epsiode name as key
data = {}
for url in episode_urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    text = [i.string for i in soup.find_all('p') if i.string not in ['Written by Chuck Lorre and Bill Prady','Advertisements','Like','Loading...','%d']]
    data[soup.find('h2').string] = pd.Series(text).drop_duplicates().tolist()
```

```python
# Dump data
with open("BBT_episodes.json","w") as file:
    json.dump(data, file, indent = 4)
```
